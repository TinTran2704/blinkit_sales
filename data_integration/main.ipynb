```python
# NOTEBOOK CELL 1 - Setup and Imports
import pandas as pd
# Your imports here
from utils import *
from datetime import datetime
#from arguments import FULL_REFRESH, INCREMENTAL_VALUE
```

```python
# NOTEBOOK CELL 2 - Configuration
SRC_PATH = "data_raw\\"
TGT_PATH = "data_cleaned\\"
```

```python
# NOTEBOOK CELL 3 - Initialize Components
tables_to_sync = get_tables_to_sync()
transform_config = read_yml_config()
```

### blinkit_customers

```python
name_job = "blinkit_customers"


start_job(name_job)
df = pd.read_csv(SRC_PATH + name_job + ".csv")

# Lấy giá trị P_KEY động theo JOB_NAME
pkey_str = tables_to_sync.loc[tables_to_sync["JOB_NAME"] == name_job, "P_KEY"].iloc[0]

# Tách thành list các cột
pkey_cols = [col.strip() for col in pkey_str.split("|")]

tar_name = tables_to_sync.loc[tables_to_sync["JOB_NAME"] == name_job, "TARGET_TABLE"].iloc[0]
# Giữ lại dòng mà tất cả các cột khóa đều notna()
df = df.dropna(subset=pkey_cols).reset_index(drop=True)

df["ETL_updated_ts"] = datetime.now()
df["ETL_DATASOURCE_NAME"] = \
        tables_to_sync.loc[tables_to_sync["JOB_NAME"] == name_job, "DATASOURCE_NUM"].iloc[0]

df = apply_transform(df,name_job,transform_config)
```

```python
HASH_KEY_PAIR = {
    "customer_key": ["ETL_DATASOURCE_NAME", "customer_id"]
}
```

```python
for key, val in HASH_KEY_PAIR.items():
    df = add_key_column(df, columns=val, new_name_col=key)
```

```python
Error_msg_lst = []
```

```python
duplicates = df[df.duplicated(subset=pkey_cols, keep=False)]
if not duplicates.empty:
    Error_msg_lst.append(f"Có {len(duplicates)} dòng bị trùng khóa p_key:")
```

```python
tar_name = tables_to_sync.loc[tables_to_sync["JOB_NAME"] == name_job, "TARGET_TABLE"].iloc[0]
df.to_csv(TGT_PATH + tar_name + ".csv", index=False)
```

```python
end_job(name_job, Error_msg_lst)
```
